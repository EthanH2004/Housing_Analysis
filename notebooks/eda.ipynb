# Texas Housing Price EDA

Use this notebook to inspect the raw and cleaned housing data.

```python
import pandas as pd

raw_path = "../ data/raw/texas_housing.csv"
cleaned_path = "../data/processed/cleaned_housing.csv"

try:
    raw_df = pd.read_csv(raw_path)
    display(raw_df.head())
except FileNotFoundError:
    print(f"Raw data not found at {raw_path}")

try:
    cleaned_df = pd.read_csv(cleaned_path)
    display(cleaned_df.describe())
except FileNotFoundError:
    print(f"Cleaned data not found at {cleaned_path}")
```
