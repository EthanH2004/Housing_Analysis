# Texas Housing Price Exploration

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

```python
# Create sample data for demonstration
sample_data = {
    'price': [250000, 300000, 275000, 320000, 290000],
    'bedrooms': [3, 4, 3, 4, 3],
    'bathrooms': [2, 2.5, 2, 3, 2],
    'square_feet': [1500, 2000, 1600, 2200, 1700]
}

df = pd.DataFrame(sample_data)
print("Sample Housing Data:")
print(df)
```

```python
# Calculate basic statistics
print("Dataset Statistics:")
print(f"Average Price: ${df['price'].mean():,.0f}")
print(f"Average Bedrooms: {df['bedrooms'].mean():.1f}")
print(f"Average Bathrooms: {df['bathrooms'].mean():.1f}")
print(f"Average Square Feet: {df['square_feet'].mean():.0f}")
```

```python
# Create a visualization
plt.figure(figsize=(10, 6))
plt.scatter(df['square_feet'], df['price'], color='steelblue', s=100, alpha=0.7)
plt.xlabel('Square Feet', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('Housing Price vs Square Footage', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Visualization complete!")
```

```python
# Display the dataframe in a nicely formatted table
print("\nDetailed Housing Data:")
print(df.to_string(index=False))

# Calculate correlation
correlation = df['square_feet'].corr(df['price'])
print(f"\nCorrelation between Square Feet and Price: {correlation:.3f}")
```
